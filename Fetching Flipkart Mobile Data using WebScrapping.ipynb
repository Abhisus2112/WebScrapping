```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
```

```python
final = pd.DataFrame()

for j in range(1,44):
    url = 'https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=1fadc05d-1d98-4780-a4bb-21e45937efb1&page={}'.format(j)
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers = headers).text
    
    soup = BeautifulSoup(webpage,"html.parser")
    product = soup.find_all('div', class_ = '_3pLy-c row')
    
    name = []
    rating = []
    no_of_rating = []
    memory = []
    display = []
    camera = []
    battery = []
    processor = []
    warenty = []
    price = []
    offer = []

    for i in product:
        name.append(i.find('div', class_ = '_4rR01T').text.strip())
        
        if (i.find('div', class_ = '_3LWZlK')) is not None:
            rating.append(i.find('div', class_ = '_3LWZlK').text.strip())
        else:
            rating.append(np.nan)
        if (i.find('span', class_ = '_2_R_DZ')) is not None:
            no_of_rating.append(i.find('span', class_ = '_2_R_DZ').text.strip())
        else:
            no_of_rating.append(np.nan)
        if (i.find_all('li', class_ = 'rgWa7D')[0]) is not None:
            memory.append(i.find_all('li', class_ = 'rgWa7D')[0].text.strip())
        else:
            memory.append(np.nan)
        if (i.find_all('li', class_ = 'rgWa7D')[1]) is not None:
            display.append(i.find_all('li', class_ = 'rgWa7D')[1].text.strip())
        else:
            display.append(np.nan)
        if (i.find_all('li', class_ = 'rgWa7D')[2]) is not None:
            camera.append(i.find_all('li', class_ = 'rgWa7D')[2].text.strip())
        else:
            camera.append(np.nan)
        if (i.find_all('li', class_ = 'rgWa7D')[3]) is not None:
            battery.append(i.find_all('li', class_ = 'rgWa7D')[3].text.strip())
        else:
            battery.append(np.nan)
        if (i.find_all('li', class_ = 'rgWa7D')[-2]) is not None:
            processor.append(i.find_all('li', class_ = 'rgWa7D')[-2].text.strip())
        else:
            processor.append(np.nan)
        if (i.find_all('li', class_ = 'rgWa7D')[-1]) is not None:
            warenty.append(i.find_all('li', class_ = 'rgWa7D')[-1].text.strip())
        else:
            warenty.append(np.nan)
        if (i.find('div', class_ = '_30jeq3 _1_WHN1')) is not None:
            price.append(i.find('div', class_ = '_30jeq3 _1_WHN1').text.strip())
        else:
            price.append(np.nan)
        if (i.find('div', class_ = '_3Ay6Sb')) is not None:
            offer.append(i.find('div', class_ = '_3Ay6Sb').text.strip())
        else:
            offer.append(np.nan)

    
    d = {'name':name, 'rating':rating,'no_of_rating':no_of_rating,'memory':memory,'display':display,'camera':camera,'battery':battery,'processor':processor, 'warenty':warenty,'price':price, 'offer':offer}    
    df1 = pd.DataFrame(d)
    
    
    final = final.append(df1,ignore_index= True)
    
    
    
```
