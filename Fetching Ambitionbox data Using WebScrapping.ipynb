## Webscrapping

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers = headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
company = soup.find_all('div', class_ = 'company-info-wrapper')
```

```python
name = []
rating = []
review = []
ctype = []
hq = []
old = []
employees = []


for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p', class_='rating').text.strip())
    review.append(i.find('a', class_='review-count').text.strip())
    ctype.append(i.find_all('p', class_ = 'infoEntity')[0].text.strip())
    hq.append(i.find_all('p', class_ = 'infoEntity')[1].text.strip())
    old.append(i.find_all('p', class_ = 'infoEntity')[2].text.strip())
    employees.append(i.find_all('p', class_ = 'infoEntity')[-1].text.strip())
    

d = {'name':name,'rating':rating,'review':review,'ctype':ctype,'hd':hq,'old':old, 'employees':employees}  

df = pd.DataFrame(d)
```

```python
final = pd.DataFrame()

for i in range(1,11):
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(i)
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers = headers).text
    
    soup = BeautifulSoup(webpage, 'lxml')
    
    company = soup.find_all('div', class_ = 'company-info-wrapper')
    
    
    name = []
    rating = []
    review = []
    ctype = []
    hq = []
    old = []
    employees = []


    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p', class_='rating').text.strip())
        review.append(i.find('a', class_='review-count').text.strip())
        ctype.append(i.find_all('p', class_ = 'infoEntity')[0].text.strip())
        ##hq.append(i.find_all('p', class_ = 'infoEntity')[-3].text.strip())
        old.append(i.find_all('p', class_ = 'infoEntity')[2].text.strip())
        employees.append(i.find_all('p', class_ = 'infoEntity')[-1].text.strip())
        
    
    

    d = {'name':name,'rating':rating,'review':review,'ctype':ctype,'old':old, 'employees': employees}  

    df1 = pd.DataFrame(d)
    
    final=final.append(df1,ignore_index = True)
```
